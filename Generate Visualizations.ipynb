```python
import warnings
warnings.simplefilter("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#create plot for training against random and rule-based agents
episodes = []
rates = []

with open('logs/sliding-random.txt') as f:
    for l in f.readlines():
        lol = l.strip().split(' ')
        episodes.append(int(lol[0]))
        rates.append(float(lol[1]))

plt.figure(figsize=(10,5))
df = pd.DataFrame({'Episode': episodes, 'Win Rate': rates})
sns.lineplot(x="Episode", y="Win Rate", data=df)
```
